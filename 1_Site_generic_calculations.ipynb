# Initialisation

```python
import bw2data as bd
import bw2io as bi
import bw2calc as bc
import pandas as pd
import bw2regional as br
import numpy as np
bd.projects.set_current("Brightway regional tutorial")
```

# Impacts calculation

```python
biodiesel_rapeseed = [activity for activity in bd.Database('BW2regional_tutorial') if "Esterification - Rapeseed - MJ" in activity['name']][0]
```

```python
FU = {biodiesel_rapeseed:1}
```

```python
lca_biodiesel = bc.LCA(FU)
lca_biodiesel.lci()
```

```python
method = [i for i in bd.methods if "IMPACT World+ Midpoint 2.1 for ecoinvent v3.10" == i[0]]
name_impacts = [i[2] for i in method] 
```

```python
name_impacts
```

```python
dict_results={}
labels=['unit','Biodiesel - Rapeseed [1 MJ]']
for i in range(len(method)):       
    data=[]
    data.append(bd.methods[method[i]]['unit'])
    lca_biodiesel.switch_method(method[i]) 
    # calculation
    lca_biodiesel.lcia()
    data.append("{:e}".format(lca_biodiesel.score))
    dict_results[name_impacts[i]]=pd.Series(np.array(data), index=labels)#result
    dfxx=pd.DataFrame(dict_results).transpose()

dfxx
```
